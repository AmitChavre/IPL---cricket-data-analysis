```python

death_overs = deliveries[deliveries['over'] >= 15]

bowler_stats = death_overs.groupby('bowler').agg({
    'total_runs': 'sum',
    'ball': 'count'
}).reset_index()

bowler_stats = bowler_stats[bowler_stats['ball'] >= 120]


bowler_stats['economy'] = (bowler_stats['total_runs'] / bowler_stats['ball']) * 6

top_bowlers = bowler_stats.sort_values('economy').head(10)

plt.figure(figsize=(10,6))

sns.barplot(
    x='economy',
    y='bowler',
    data=top_bowlers,
    palette='coolwarm'
)

plt.title('Top 10 Economical Bowlers in Death Overs (16–20)')
plt.xlabel('Economy Rate')
plt.ylabel('Bowler')

plt.tight_layout()
plt.show()
```

## SP Narine and DE Bollinger emerge as the most economical bowlers in death overs, consistently restricting runs under pressure.
## The analysis also highlights that both spinners and pacers play a crucial role in controlling scoring during the final overs.

```python
# Step 1: Create empty column
matches['bat_first_team'] = ''

# Step 2: If toss decision = bat
mask_bat = matches['toss_decision'] == 'bat'
matches.loc[mask_bat, 'bat_first_team'] = matches.loc[mask_bat, 'toss_winner']

# Step 3: If toss decision = field
mask_field = matches['toss_decision'] == 'field'

# Case 1: toss_winner == team1 → team2 bats first
mask_team1 = mask_field & (matches['toss_winner'] == matches['team1'])
matches.loc[mask_team1, 'bat_first_team'] = matches.loc[mask_team1, 'team2']

# Case 2: toss_winner == team2 → team1 bats first
mask_team2 = mask_field & (matches['toss_winner'] == matches['team2'])
matches.loc[mask_team2, 'bat_first_team'] = matches.loc[mask_team2, 'team1']

# Step 4: Create win_type column
matches['win_type'] = 'chase'
matches.loc[matches['winner'] == matches['bat_first_team'], 'win_type'] = 'bat_first'

# Step 5: Count wins
win_counts = matches.groupby(['winner', 'win_type']).size().unstack(fill_value=0)

# Step 6: Top teams
top_teams = win_counts.sum(axis=1).sort_values(ascending=False).head(10).index
win_counts = win_counts.loc[top_teams]

# Step 7: Plot
win_counts.plot(kind='bar', figsize=(12,6), colormap='Set2')

plt.title('Bat First vs Chasing Wins (Top Teams)')
plt.xlabel('Teams')
plt.ylabel('Wins')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

## Most teams show a higher number of wins while chasing, indicating a general advantage in pursuing targets.
## Kolkata Knight Riders and Rajasthan Royals stand out as strong chasing teams, while Mumbai Indians remain balanced in both strategies.

```python
# Step 1: Filter death overs
death_overs = deliveries[deliveries['over'] >= 15]

# Step 2: Create bat_stats
bat_stats = death_overs.groupby('batter').agg({
    'batsman_runs': 'sum',
    'ball': 'count'
}).reset_index()

# Step 3: Filter minimum balls
bat_stats = bat_stats[bat_stats['ball'] >= 60]

# Step 4: Calculate strike rate
bat_stats['strike_rate'] = (bat_stats['batsman_runs'] / bat_stats['ball']) * 100

# Step 5: Top players
top_batters = bat_stats.sort_values('strike_rate', ascending=False).head(10)

# Step 6: Plot
sns.barplot(x='strike_rate', y='batter', data=top_batters)


plt.title('Top 10 Strike Rates in Death Overs (16–20)')
plt.xlabel('Strike Rate')
plt.ylabel('Batter')

plt.tight_layout()
plt.show()
```

## T Stubbs and AB de Villiers stand out as the most explosive finishers, maintaining exceptionally high strike rates in death overs.
